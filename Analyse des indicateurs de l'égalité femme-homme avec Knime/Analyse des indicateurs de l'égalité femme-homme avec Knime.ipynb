```python
# On recherche les doublons
print(len(df_transaction))
print(len(df_transaction.drop_duplicates()))
```

```python
# On crée un data frame avec les doublons
df_transactionDuplicate = df_transaction.loc[df_transaction.duplicated() == True]
df_transactionDuplicate.head(10)
```

```python
#On cherche des informations sur ces valeurs de "date" débutant par "test"
df_transactionTest = df_transaction[df_transaction['date'].str.contains('test')]
df_transactionTest
```

```python
# On indique à la colonne "date" que les données sont de type datetime
df_transaction['date'] = pd.to_datetime(df_transaction['date'])
print(df_transaction.info())
```

```python
# On joint df_produits et df_transaction sur la colonne "id_prod" pour commencer notre dataframe globale
df_globale = pd.merge(df_transaction, df_produits, on = 'id_prod') 
print(df_globale.head())
print('- - - - - - - - - - - - - - - ')
print(df_globale.info())
```

```python
# On vérifie s'il y a d'improbables doublons
print(len(df_globale))
print(len(df_globale.drop_duplicates()))
```

```python
#Nous regardons pourquoi toutes les transactions n'apparaissent pas dans notre df_globale
df_transactionGroupéIDProd = df_transaction.groupby('id_prod').count().reset_index()
print(df_transactionGroupéIDProd.head())
print('- - - - - - - - - - - - - -')
for i in df_globale.groupby('id_prod').count().reset_index()['id_prod']:
            df_transactionGroupéIDProd.drop(df_transactionGroupéIDProd.loc[df_transactionGroupéIDProd['id_prod'] == i].index, inplace=True)
print(df_transactionGroupéIDProd)
```

```python
X = "âge" # qualitative
Y = "nbeAchatProduit" # quantitative

sous_echantillon = df_achatClients.loc[df_achatClients['categ'] == '0']

def eta_squared(x,y):
    moyenne_y = y.mean()
    classes = []
    for classe in x.unique():
        yi_classe = y[x==classe]
        classes.append({'ni': len(yi_classe),
                        'moyenne_classe': yi_classe.mean()})
    SCT = sum([(yj-moyenne_y)**2 for yj in y])
    SCE = sum([c['ni']*(c['moyenne_classe']-moyenne_y)**2 for c in classes])
    return SCE/SCT
    
eta_squared(sous_echantillon[X],sous_echantillon[Y])
```
